```python
import os
import pandas as pd
from tqdm import tqdm  # 导入tqdm模块

deg_gene = 'drive/MyDrive/BS6202/DEG_genelist_P-adj.csv'
deg_gene_df = pd.read_csv(deg_gene)

# 读取phenotype文件
phenotype_file = 'drive/MyDrive/BS6202/phenotype144.xlsx'
phenotype_data = pd.read_excel(phenotype_file)

# 创建一个用于存储筛选数据的DataFrame
TCGA_GC = pd.DataFrame()

# 使用tqdm创建进度条
for index, row in tqdm(phenotype_data.iterrows(), total=len(phenotype_data)):
    file_id = row['File_ID']
    sample_type = row['Sample_Type']

    # 构建基因表达数据文件路径
    data_folder = os.path.join('drive/MyDrive/BS6202/gdc_download_20231012_104140.953493', file_id)
    data_file = os.path.join(data_folder, 'rna_seq.augmented_star_gene_counts.tsv')

    try:
        # 读取基因表达数据
        gene_expression_data = pd.read_csv(data_file, sep='\t', header=1)

        # 检查是否存在'gene_type'列
        if 'gene_type' in gene_expression_data.columns:
            # 筛选出gene_type为"protein_coding"的数据
            gene_expression_data = gene_expression_data[gene_expression_data['gene_id'].isin(deg_gene_df['gene_id'])]

            # 添加新列 'Sample_Type' 和 'File_ID'
            gene_expression_data.loc[:, 'Sample_Type'] = sample_type
            gene_expression_data.loc[:, 'File_ID'] = file_id

            # 将数据添加到TCGA_GC DataFrame中
            TCGA_GC = pd.concat([TCGA_GC, gene_expression_data], ignore_index=True)
        else:
            print(f"'gene_type' column not found in sample {file_id}. Skipping.")
    except KeyError:
        print(f"KeyError encountered for sample {file_id}. Skipping.")

# TCGA_GC现在包含了所有符合条件的基因表达数据，并包括新的 'Sample_Type' 和 'File_ID' 列
```
